## 4. for文
正方形を描く例題、一辺を描画して向きを変える操作を4回も書くのは大変だったのではないでしょうか？「同じコードを書くなら何回もコピペしちゃえ」という楽もできるとは思います。しかしプログラミングには、この「何回もコピペする」ことも楽して書くために、`for`文という機能があります。<br><br>

下のブロックを実行してみてください。`for`の下2行には、

- `forward(200)`で真っすぐ線を描く
- `left(90)`で左に90度向く (と真上を向くようになる)

という処理しか書かれていません。`for`文とは、その下に半角スペース4つ分 (「インデント」と呼びます。`Tab`キーでスペース4つ分を一気に入力できます。)空けて書いた処理を繰り返す文法です。`for i in range(4)`と書くと、4回繰り返すことができます。

```python
turtle.setup(width=500, height=500)
for i in range(4):
    turtle.forward(200)
    turtle.left(90)
turtle.done()
```

## 5. 関数
この課題では10行程度のソースコードですが、テトリスのソースコードは数百行に及びます。似たような処理をするために`N`の値を変えたいとき、数百行のソースコードから`N`を探し出して値を変える・・・というのは、骨が折れる作業だと思います。そこで、処理をひとまとめにすることのできる「関数」が便利です。<br><br>

下のブロックを実行してみてください。関数を使うには2つのステップがあり、まず関数の名前とそのひとまとめにしたい処理を「定義」してあげます。このブロックは定義だけなので、何も表示されません。

```python
def drawRegularNgon(N):
    if N<=0:
        print("Invalid value!")
        return

    turtle.setup(width=500, height=500)
    for i in range(N):
        turtle.forward(50)
        turtle.left(360/N)
    turtle.done()
```

次に、関数に値を指定して「コール」します。括弧の中の数字を自由に変更した上で、下のブロックを実行してみてください。このコードだけでどのような正N角形も描けます。括弧の中身のことを、関数の「引数」と呼びます。

```python
drawRegularNgon(10)
```

さらに、長さも指定したい場合にも対応できるように引数を増やしてみましょう。関数を定義したブロックと、関数をコールするブロックを順に実行してみてください。そして、その引数を自由に変更してみてください。

```python
def drawRegularNgon(N, length=50):
    if N<=0:
        print("Invalid value!")
        return

    turtle.setup(width=500, height=500)
    for i in range(N):
        turtle.forward(length)
        turtle.left(360/N)
    turtle.done()
```

```python
drawRegularNgon(10, 100)
```

## 6. 再帰関数 (おまけ)
ある関数の中で自分自身をコールする方法を、「再帰関数」と呼びます。「再帰(=深さ優先探索、DFS)」は、テトリスで言えば「3手先まで考える」場合に使われます。現時点では知っておく程度で大丈夫です。

```python
def drawMountainLine(length,dim):
    if dim == 0:
        turtle.forward(length)
        turtle.left(60)
        turtle.forward(length)
        turtle.right(120)
        turtle.forward(length)
        turtle.left(60)
        turtle.forward(length)
    else:
        drawMountainLine(length/3,dim-1)
        turtle.left(60)
        drawMountainLine(length/3,dim-1)
        turtle.right(120)
        drawMountainLine(length/3,dim-1)
        turtle.left(60)
        drawMountainLine(length/3,dim-1)

turtle.tracer(False) #亀ちゃんの軌跡を追うと重くなるため、今回は追わない設定に
turtle.setup(width=1500, height=700)
turtle.up()
turtle.backward(500)
turtle.down()

drawMountainLine(300,4)

turtle.done()
```
